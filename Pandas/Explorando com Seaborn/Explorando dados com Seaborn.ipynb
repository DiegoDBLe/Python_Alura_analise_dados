# Importando Base de Dados

```python
import pandas as pd
```

## Traduzindo o nome das colunas para português

```python
renomear = {
    'total_bill' : 'valor_da_conta',
    'tip' : 'gorjeta',
    'dessert' : 'sobremesa',
    'day' : 'dia_da_semana',
    'time' : 'hora_do_dia',
    'size' : 'total_de_pessoas'
}
```

### Alterando o nome das colunas e atribuindo a uma nova variavel

```python
gorjeta = dados.rename(columns= renomear)
```

```python
gorjeta.head()
```

### Alterando as informações de inglês para português de cada linha
- Primeiro utilizando o método unique(), retorna quais os tipos de informaçõe tem em cada coluna que chamamos

```python
gorjeta['sobremesa'].unique()
```

```python
sim_nao = {
    'No' : 'Não',
    'Yes' : 'Sim'
}
```

#### Mapeando na coluna sobremesa os No e Yes para sim e Não

```python
gorjeta['sobremesa'].map(sim_nao)
```

### Atribuindo a variavel gorgeta, coluna sobremesa a alteração 

```python
gorjeta['sobremesa'] = gorjeta['sobremesa'].map(sim_nao)
```

```python
gorjeta.sample(5)
```

## Alterando o nome dos dias da semana

```python
gorjeta['dia_da_semana'].unique()
```

```python
dias_semana = {
    'Sun' : 'Domingo',
    'Sat' : 'Sábado',
    'Thur' : 'Quinta-feira',
    'Fri' : 'Sexta-feira'
}
```

```python
gorjeta['dia_da_semana'] = gorjeta['dia_da_semana'].map(dias_semana)
```

```python
gorjeta.sample(10)
```

## Alterando os tipos de refeição

```python
gorjeta['hora_do_dia'].unique()
```

```python
tipo_refeicao = {
    'Dinner' : 'Jantar',
    'Lunch' : 'Almoço'
}
```

```python
gorjeta['hora_do_dia'] = gorjeta['hora_do_dia'].map(tipo_refeicao)
```

```python
gorjeta.sample(10)
```

## Continuando a análise: Criando uma variavel(coluna) chamada de procentagem
- Onde nessa coluna iremos colocar o percentual de cada gorjeta em relação ao valor da conta
- Iremos analisar se o valor da gorjeta é proporcional ao valor da conta
- Para saber o percentual é só caucular a gorjeta / valor da conta

```python
gorjeta['porcentagem'] = gorjeta['gorjeta'] / gorjeta['valor_da_conta']
```

```python
gorjeta.head(5)
```

```python
# Arredondando os valores da coluna porcentagem:
gorjeta['porcentagem'] = gorjeta['porcentagem'].round(2)
```

```python
gorjeta.head(5)
```

```python
porcentagem_conta = sns.scatterplot(x = 'valor_da_conta', y = 'porcentagem', data = gorjeta)
```

### Visualmente o valor da conta não é proporcional ao valor da gorjeta
