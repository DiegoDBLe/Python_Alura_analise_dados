```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15, 8))
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
```

```python
dados.head(6)
```

### Criando uma area para colocar quatro gráficos 

### Criando a area:

```python
area = plt.figure()
```

### Criando os quatros Gráficos
 - COnfigurando os gráficos

```python
# To querendo dizer que dentro dessa figura terão 4 gráficos duas linhas e duas colunas e esse primeiro vai ficar na posição
# 1
grafico_1 = area.add_subplot(2, 2, 1)
grafico_2 = area.add_subplot(2, 2, 2)
grafico_3 = area.add_subplot(2, 2, 3)
grafico_4 = area.add_subplot(2, 2, 4)
```

### Começar a fazer os gráficos:

```python
grafico_1.scatter(dados['Valor'], dados['Area'])
grafico_1.set_title('Valor x Area')

grafico_2.hist(dados['Valor'])
grafico_2.set_title('Histograma')

dados_grafico_3 = dados['Valor'].sample(100)
dados_grafico_3.index = range(dados_grafico_3.shape[0])
grafico_3.plot(dados_grafico_3)
grafico_3.set_title('Amostra (Valor)')

grupo = dados.groupby('Tipo')['Valor']
label = grupo.mean().index
valores = grupo.mean().values
grafico_4.bar(label, valores)
grafico_4.set_title('Valor Médio por Tipo')
```
