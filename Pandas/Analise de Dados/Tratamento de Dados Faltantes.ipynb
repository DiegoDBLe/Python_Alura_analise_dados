### Tratamento de Dados Faltantes

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial.cvs', sep = ';')
```

```python
dados.head(10)
```

#### Método .isnull():
- Ele cria um DataFrame com os valores onde é nullo com o valor True, ou seja, onde for nullo ele mostra True

```python
dados.isnull()
```

#### Método .notnull():
- Ele cria um DataFrame com os valores onde é nullo com o valor False, ou seja, onde for nullo ele mostra False, esse método é o contrário do isnull()

```python
dados.notnull()
```

#### Métodos .info():
- Ele retorna as informações do DataFrame, quantidade de colunas, index da coluna, quantidade de linhas, tipos das colunas e quantidade de non-null, ou seja, numeros nullos.

```python
dados.info()
```

```python
# Dessa forma ele retorna uma Series com valores booleanos e onde for Nullo ele retorna True
dados.Valor.isnull()
```

```python
# Passando nossa Series para o DataFrame: Dessa maneira ele retorna o DataFrame com todos os valores nullos da coluan Valor
dados[dados.Valor.isnull()]
```

```python
# Quantos valores nullos possui na coluna valor: usando o método len()
len(dados[dados.Valor.isnull()])
```

```python
# Quantos valores nullos possui na coluna valor: usando o método .shape[0]
dados[dados.Valor.isnull()].shape[0]
```

```python
# Apagando os valores nullos da coluna Valor, pois nesse exemplo não servem para nada:
dados.dropna(subset= ['Valor'], inplace= True)
```

```python
# Conferindo se ainda tem valor nullo na coluna Valor:
dados[dados.Valor.isnull()]
```

### Tratamento de Dados Faltantes (Continuação)

```python
# Vamor tratar os dados nullos da coluna Condominio:
dados[dados['Condominio'].isnull()]
```

```python
# Quantidade de imóveis com Condominio Nan:
len(dados[dados['Condominio'].isnull()])
```

```python
# Pegar os Tipos de imóveis apartamento com o valor do condomino NaN:
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```python
selecao
```

```python
# Mostrando no DataFrame os apartamentos com condomino NaN:
dados[selecao].head(30)
```

```python
# Quantidade de Apartamento com condominio NaN:
len(dados[selecao])
```

```python
# A variavel selecao, seleciona na coluna Tipo somente os apartamentos e na coluna Condominio os nullos, ou seja retorna
# Uma series com os valores True quando for nulo. Aqui estamos sobrescrevendo o DataFrame dados sem os Apartametos com 
# Condominio Nullos, pois (~) ele inverte a series booleana, entao onde era True fica False e onde era False fica True,
# sendo assim o novo DataFrame fica sem os 745 registros de apartamentos com Condominio NaN
dados = dados[~selecao]
```

```python
dados.head(50)
```

```python
# "novo DataFrame" sem os apartamentos com condominio NaN:
dados.shape[0]
```

```python
# Agora ficou quantos com condominio nullos?
dados[dados['Condominio'].isnull()].shape[0]
```

```python
# Vamos manter o restante dos condominios nullos e o IPTU nullo também:
# Vamos atribuir o valor 0 para esses valores nullos, usando a função .
# #xistem duas formas de substiruir esses valores:
# 1° dados.fillna(0, inplace=True) -> Aqui substiruiria todos os valores NaN por 0.
# E da forma abaixo onde atribuo a uma variavel o DataFrame com o método .fillna() e passo como parametro
# um dicionário onde posso escolher a coluna que quero alterar e o valor que quero colocar:
dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
```

```python
dados.head(20)
```

```python
# Confirmando se tem numero nullo na coluna Condominio:
dados[dados['Condominio'].isnull()].shape[0]
```

```python
# Confirmando se tem numero nullo na coluna IPTU:
dados[dados['IPTU'].isnull()].shape[0]
```

```python
len(dados)
```

```python
dados.info()
```

```python
# Salvando o Arquivo:
dados.to_csv('aluguel_residencial.cvs', sep = ';', index = False)
```

# Exercicios:

```python
imoveis = pd.DataFrame([['Apartamento', None, 970, 68], 
                        ['Apartamento', 2000, 878, 112], 
                        ['Casa', 5000, None, 500], 
                        ['Apartamento', None, 1010, 170], 
                        ['Apartamento', 1500, 850, None], 
                        ['Casa', None, None, None], 
                        ['Apartamento', 2000, 878, None], 
                        ['Apartamento', 1550, None, 228], 
                        ['Apartamento', 2500, 880, 195]], 
                        columns = ['Tipo', 'Valor', 'Condominio', 'IPTU'])
```

```python
imoveis
```

#### 1) Elimina os registros que não apresentam a variável Valor:

```python
imoveis.dropna(subset=['Valor'], inplace= True)
```

```python
imoveis
```

#### 2) Elimina os imóveis do tipo Apartamento que não apresentam valor Condominio:

```python
selecao = (imoveis['Tipo'] == 'Apartamento') & (imoveis['Condominio'].isnull())
```

```python
imoveis = imoveis[~selecao]
```

```python
imoveis
```

####  Substitui os valores faltantes que restam nas variáveis Condominio e IPTU por zero:

```python
imoveis = imoveis.fillna({'Condominio':0, 'IPTU':0})
```

```python
imoveis
```

#### 4) Reconstrói o índice do DataFrame resultante:

```python
imoveis.index = range(len(imoveis))
```
