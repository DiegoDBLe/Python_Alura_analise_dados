## Criando Novas Variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial.cvs', sep = ';')
```

```python
dados.head(10)
```

#### Criando uma nova variável chamada de Valor Bruto:
- Valor Bruto é a soma da coluna Valor + Condominio + IPTU

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados
```

#### Criando uma nova variável chamada de Valor m²:
- Valor m² é a divisão da coluna Valor / coluna Area

```python
dados['Valor m²'] = dados['Valor'] / dados['Area']
```

```python
dados
```

```python
# Arredondando em duas casas decimais a coluna Valor m²:
dados['Valor m²'] = dados['Valor m²'].round(2)
```

```python
dados.head(10)
```

#### Criando uma nova variável chama da Valor Bruto m²:
- Valor Bruto m² é coluna Valor Bruto / coluna Area

```python
dados['Valor Bruto m²'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.head(15)
```

#### Criando uma nova variável para especificar o tipo de imovel:
- Exemplo: Casa, Casa de Condominio, e Casa de vila serão do tipo casa. Já os demais serão do tipo apartamento

```python
# 1° Criar um listar com os tipos de casa:
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
casa
```

```python
# criaremos efetivamente a nova variável que chamaremos de Tipo Agregado. Depois, evocaremos dados e passaremos a variável 
# Tipo, seguido do método apply(). Esse método pertmite que apliquemos uma função à cada registro do DataFrame. Para tanto,
# criaremos uma função lambda : 'lambda x: 'Casa' if x in casa else 'Apartamento'

dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados.head(20)
```

# Excluindo Variáveis:
- 1° Vou criar outro DataFrame so para testar antes de apagar do DataFrame Dados.

##### Criando um novo DataFrame para fazer teste de excluir variáveis:

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m²', 'Valor Bruto', 'Valor Bruto m²']])
dados_aux.head(15)
```

####  1° Forma de apagar uma variável: usando o del

```python
del dados_aux['Valor Bruto']
```

```python
dados_aux.head(15)
```

#### 2° Forma de apagar uma variável: usando o pop()

```python
dados_aux.pop('Valor Bruto m²')
```

```python
dados_aux
```

#### # 3° Forma de apagar uma variável é utilizando o drop([]) -> podemos passar uma lista com o nome das colunas que queremos apagar, e o argumento axis = 1.   OBS: axis = 0 é para linhas e axis = 1 é para colunas.

```python
dados_aux.drop(['Tipo Agregado', 'Valor m²'], axis= 1, inplace = True )
```

```python
dados_aux.head(15)
```

### Apagando do DataFrame dados:

```python
dados.drop(['Valor Bruto','Valor Bruto m²'], axis = 1, inplace = True)
```

```python
dados.head(15)
```

```python
dados.to_csv('aluguel_residencial.csv', sep = ';', index = False)
```
