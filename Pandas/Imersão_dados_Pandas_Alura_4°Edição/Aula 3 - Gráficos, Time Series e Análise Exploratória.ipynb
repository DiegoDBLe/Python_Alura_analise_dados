## Preço do metro quadrado por bairro e plotar em um gráfico ideal:

```python
dados_vendas.head(10)
```

```python
# Criando a coluna 'Valor m²' e já atribuindo a ela o valor do m²

dados_vendas['Valor_m²'] = dados_vendas['Valor_anuncio_float'] / dados_vendas['Metragem']
dados_vendas
```

```python
# Agrupando e tirando uma 'Média' das metragens, quartos, banheiros, vagas, Valor_anuncio_float, Valor_m²
dados_vendas.groupby('Bairro').mean()
```

```python
# somando Metragem  	 Quartos 	Banheiros	Vagas	Valor_anuncio_float 	 Valor_m² por bairro:
dados_bairro = dados_vendas.groupby('Bairro').sum()
dados_bairro.head(10)
```

```python
# Criando uma variavel(Coluna) 'Valor_m²_bairro' e atribuindo a ela o valor do m² de cada bairro
dados_bairro['Valor_m²_bairro'] = dados_bairro['Valor_anuncio_float'] / dados_bairro['Metragem']
dados_bairro
```

```python
# Localizando pelo nome do bairro:
dados_bairro.loc['Vila Mariana']
```

```python
plt.figure(figsize=(16, 14))
ax = sns.barplot(x='Bairro', y='Valor_m²_bairro', data=dados_bairro.query('Bairro in @dez_primeiros_bairros'))
```

```python
plt.figure(figsize=(16, 14))
ax = sns.boxplot(data = dados_vendas.query('Bairro in @dez_primeiros_bairros'), x='Bairro', y='Valor_anuncio_float')
plt.show()
```
